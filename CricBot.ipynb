```python
def overview(team):
    resp=requests.get('https://www.cricbuzz.com/')
    soup = BeautifulSoup(resp.content, 'html.parser')
    All_matches=soup.find_all('li',attrs = {'class': 'cb-col cb-col-25 cb-mtch-blk cb-vid-sml-card-api videos-carousal-item cb-carousal-item-large cb-view-all-ga'})
    for match in All_matches:
        team_score=match.text.strip()
        if team_score.count(team)>0:
            aTag_attr=match.find('a',attrs = {'class': 'cb-font-12'})
            link=aTag_attr['href']
            
            return link,team_score

def fetch_scores(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html5lib')
    stats=soup.find('div',class_='cb-col-67 cb-col')
    #here we have tofindout by name
    if stats==None:
        return "Match is Over \n"
    scoreD=stats.find_all('div',class_='cb-min-inf cb-col-100')
    s1=scoreD[0]
    s2=scoreD[1]
#     print(s1.prettify())
    bat=s1.find_all('div',class_='cb-col cb-col-100 cb-min-itm-rw');
    scr=""
    for batsman in bat:
        scr=scr+"\n"
        for index,score in enumerate(batsman):
            #scr=scr+score.text+"  ";
            if index==0:
                 scr=scr+score.text+":";
            elif index==2:
                scr=scr+"("+score.text+")"+" "
            elif index==5:
                 scr=scr+"SR: "+score.text
            elif index==3 or index==4:
                continue
            else:
                scr=scr+score.text+" ";
    scr=scr+"\n"          
    bowl=s2.find_all('div',class_='cb-col cb-col-100 cb-min-itm-rw');
    for bowler in bowl:
        scr=scr+"\n"
        for index,score in enumerate(bowler):
            if index==0:
                scr=scr+score.text+":"
            elif index==1:
                scr=scr+score.text+"("+"O"+") "
            elif index==3:            
                scr=scr+score.text+"("+"R"+") " 
            elif index==4:
                 scr=scr+score.text+"("+"W"+") " 
            elif index==5:
                 scr=scr+"  ECO:"+score.text 
            else:
                continue
    return scr;
    

def sendMsg(rec,msg,browser):
    try:
        search_bar = browser.find_elements(By.XPATH, '//div[@contenteditable = "true"]')[0]
        search_bar.send_keys(rec)
        search_bar.send_keys(Keys.ENTER)
        msg_bar = browser.find_elements(By.XPATH, '//div[@contenteditable = "true"]')[1]
        msgs = msg.split('\n')
        for m in msgs:
            msg_bar.send_keys(m)
            msg_bar.send_keys(Keys.SHIFT + Keys.ENTER)
            msg_bar.send_keys(Keys.SHIFT + Keys.ENTER)
        msg_bar.send_keys(Keys.ENTER)
        return 'Scorecard Sent Successfully'
    except Exception as e:
        print('Error: ', e)
        return 'Error While Sending Scorecard'

    
    


# link,ovrvw=overview('SL')
# score=fetch_scores('https://www.cricbuzz.com'+link)
# score=ovrvw+"\n"+score
# print(score)
CHROME_PROFILE_PATH="user-data-dir=C:\\Users\\user\\AppData\\Local\\Google\\Chrome\\User Data\\AutoWtsp"
options=webdriver.ChromeOptions()
options.add_argument(CHROME_PROFILE_PATH)
browser=webdriver.Chrome(executable_path='C:/Users/user/chromedriver',options=options)
browser.get("https://web.whatsapp.com/")
time.sleep(10)
batter = False

msgStatus=sendMsg('Jasi New',score,browser)
print(msgStatus)
```
